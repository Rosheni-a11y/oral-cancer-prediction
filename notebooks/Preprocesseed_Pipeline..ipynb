# Member 2 - Removal of Outliers


```python
# Visualize Age distribution using a boxplot to identify outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()

# Visualize Tumor Size distribution using a boxplot to identify outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Tumor Size (cm)'])
plt.title("Boxplot of Tumor Size")
plt.show()
```

```python
# Function to remove outliers based on IQR (Interquartile Range)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
     # Keep only rows within bounds
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Remove outliers from Age and Tumor Size columns
df = remove_outliers(df, 'Age')
df = remove_outliers(df, 'Tumor Size (cm)')

print("New shape after outlier removal:", df.shape)
```

```python
# Visualize Age after outlier removal
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()

# Visualize Tumor Size after outlier removal
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Tumor Size (cm)'])
plt.title("Boxplot of Tumor Size")
plt.show()
```

# Member 3 - Encoding categorical features

```python
# Encode binary features to numeric values
binary_cols = ['Tobacco Use', 'Alcohol Consumption', 'HPV Infection',
               'Betel Quid Use', 'Chronic Sun Exposure', 'Poor Oral Hygiene',
               'Family History of Cancer', 'Compromised Immune System',
               'Oral Lesions', 'Unexplained Bleeding', 'Difficulty Swallowing',
               'White or Red Patches in Mouth', 'Early Diagnosis', 'Gender', 'Oral Cancer (Diagnosis)']

# Map 'Yes'/'No' and 'Male'/'Female' to 1/0
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0, 'Male':1, 'Female':0})

# Encode ordinal features
# Convert Diet (Fruits & Vegetables Intake) to numeric: Low=0, Moderate=1, High=2
diet_mapping = {'Low':0, 'Moderate':1, 'High':2}
df['Diet (Fruits & Vegetables Intake)'] = df['Diet (Fruits & Vegetables Intake)'].map(diet_mapping)

# Target encoding for Country feature
# Compute mean target value for each country
country_mean = df.groupby('Country')['Oral Cancer (Diagnosis)'].mean()

# Replace Country with its encoded mean value
df['Country_encoded'] = df['Country'].map(country_mean)

# Drop original Country column as it's now encoded
df.drop('Country', axis=1, inplace=True)  

# Show first few rows after encoding
df.head()


```

```python
# Visualize Tobacco Use categorical feature before encoding and after encoding
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before encoding
sns.countplot(x=df['Tobacco Use'].replace({1:'Yes', 0:'No'}), ax=axes[0])
axes[0].set_title("Before Encoding (Yes/No)")

# After encoding
sns.countplot(x=df['Tobacco Use'], ax=axes[1])
axes[1].set_title("After Encoding (1/0)")

plt.tight_layout()
plt.show()
```

# Member 4 - Feature Engineering

```python
# Create a correlation matrix for all features
plt.figure(figsize=(12,8))
corr = df.corr()

# Plot a heatmap of correlations
sns.heatmap(corr, annot=False, cmap='coolwarm', cbar=True, linewidths=0.5)
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()

# Get correlation of each feature with the target variable
target_corr = corr['Oral Cancer (Diagnosis)'].sort_values(ascending=False)
print("Correlation with Oral Cancer (Diagnosis):")
print(target_corr)
```

```python
# Step 1: Drop leakage/highly correlated features features
df = df.drop(columns=['Cancer Stage', 'Tumor Size (cm)','Country_encoded'])
df.head()
```

```python
#step 2 : Creating new features
# 1. Combined Risk Features
# High_Risk_Habits: sum of lifestyle habits that increase oral cancer risk
df['High_Risk_Habits'] = df['Tobacco Use'] + df['Alcohol Consumption'] + df['Betel Quid Use']

# Oral_Health_Risk: sum of oral health related indicators
df['Oral_Health_Risk'] = (df['Poor Oral Hygiene'] + df['Oral Lesions'] + df['Difficulty Swallowing'] + df['White or Red Patches in Mouth'] + df['Unexplained Bleeding'])

# Immune_Risk: sum of immune-compromising conditions
df['Immune_Risk'] = df['Compromised Immune System'] + df['HPV Infection']
```

```python
#2. Interaction Features
# Feature to capture interaction effect between Tobacco and Alcohol use
df['Tobacco_Alcohol_Interaction'] = df['Tobacco Use'] * df['Alcohol Consumption']

```

```python
# 3. Aggregated Count Features
# Total_Risks: total number of risks for a patient from multiple binary features
binary_features = ['Tobacco Use','Alcohol Consumption','HPV Infection','Betel Quid Use',
                   'Poor Oral Hygiene','Oral Lesions','Difficulty Swallowing',
                   'White or Red Patches in Mouth','Unexplained Bleeding','Early Diagnosis']
df['Total_Risks'] = df[binary_features].sum(axis=1)
```

```python
# step 5 - Feature Selection using Random Forest (Embedded Method)


from sklearn.ensemble import RandomForestClassifier

# Step 1: Prepare features and target
X = df.drop('Oral Cancer (Diagnosis)', axis=1)


y = df['Oral Cancer (Diagnosis)']   # Target
```

```python
#Train a Random Forest to get feature importance
# Initialize Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit model
rf.fit(X, y)

# Get feature importances
importances = pd.Series(rf.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=False)

print("Feature importances:")
print(importances)
```

```python
# step 6 - Visualize feature importance

plt.figure(figsize=(10,6))
importances.plot(kind='bar')
plt.title("Feature Importance from Random Forest")
plt.show()
```

```python
# Step 6: Select top features based on importance

# Select features with importance greater than 0.01 (arbitrary threshold)
selected_features = importances[importances > 0.01].index

# Create a new DataFrame with only the selected features
X_selected = X[selected_features]

print("Selected features:", selected_features)

X_selected.head()
```

# Member 6 - PCA/ Dimension Reduction

```python
# Apply PCA for Dimension Reduction
from sklearn.decomposition import PCA

# Initialize PCA with dynamic number of components based on total features
n_components = min(X_scaled_all.shape[1], 16) 
print("Using n_components:", n_components)

# Initialize PCA
pca = PCA(n_components=n_components)

# Fit PCA on scaled all numeric features and transform
X_pca = pca.fit_transform(X_scaled_all)

# Explained variance ratio for each principal component
# This shows how much variance each component captures
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio for each component:\n", explained_variance)

# Cumulative explained variance
# Helps decide how many components to retain most of the information
cumulative_variance = explained_variance.cumsum()
print("\nCumulative explained variance:\n", cumulative_variance)
```

```python
#  EDA Visualization for PCA 
components = np.arange(1, len(explained_variance)+1)

plt.figure(figsize=(8,5))
plt.bar(components, explained_variance, alpha=0.7, color='skyblue')
plt.plot(components, cumulative_variance, marker='o', color='red')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Scree Plot & Cumulative Explained Variance')
plt.xticks(components)
plt.grid(axis='y')
plt.legend(['Cumulative Variance', 'Individual Variance'])
plt.show()
```

```python
# Step 2: Apply PCA with selected number of components

# Decide number of components based on cumulative variance (~98.1% here)
n_components = min(X_scaled_all.shape[1], 15)  
print("Using n_components:", n_components)

# Initialize PCA with chosen components
pca = PCA(n_components=n_components)

# Transform the scaled all numeric features
X_pca_reduced = pca.fit_transform(X_scaled_all)

# Convert back to DataFrame for easier use
X_pca_df = pd.DataFrame(X_pca_reduced, 
                        columns=[f'PC{i+1}' for i in range(n_components)])

# Check shape and first few rows
print("Shape after PCA:", X_pca_df.shape)
X_pca_df.head()


```

```python
# 2D Scatter Plot (PC1 vs PC2)

plt.figure(figsize=(7,5))
sns.scatterplot(
    x=X_pca_df['PC1'], 
    y=X_pca_df['PC2'], 
    hue=y,  # your target variable
    palette='Set1',  # nice contrasting colors
    alpha=0.7,
    s=60
)
plt.title('PCA: PC1 vs PC2')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend(title='Oral Cancer', loc='best')
plt.grid(True)
plt.show()
```
